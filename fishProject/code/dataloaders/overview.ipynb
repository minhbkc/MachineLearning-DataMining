# Problem to solve
## Problem
    Based on 5 featured of 1 fish objects, predict the types of that fish
## Column information / Features:
    1. Weight: weight of fish in Gram
    2. Lenght1: vertical length in cm
    3. Length2: diagonal length in cm
    4. Length3: cross length in cm
    5. Height: height in cm
    6. Width: width in cm
    7. Species: species name of fish
## Goal
    Predict of probability of a type of fish

```python
# import data and library
import numpy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
path = '../../dat/Fish.csv'
data = pd.read_csv(path)
data.head()
```

```python
data.info(verbose=True)
```

```python
data.nunique()
```

```python
# see the distribution of numerical attributes
data.describe()
```

```python
# See if there is any null in our data
null_info = pd.DataFrame({'Null count': data.isnull().sum(), 'Null ratio': data.isnull().sum()/len(data)})
null_info.drop('Species', inplace=True)
null_info.sort_values(by='Null ratio', ascending=False, inplace=True)
null_info
```
